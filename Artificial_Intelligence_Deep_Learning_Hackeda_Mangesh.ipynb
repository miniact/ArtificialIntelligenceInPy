#Actual Implementation Of Sigmoid NN


**Working With Sigmoid Neuron:**
1. sum of dot product of input*weigths 
 Σ(wi * xi)
2. threshold value/ bias value b:  z= ( Σ(wi * xi)-b)
3. perform sigmoid function to convert it between 0 to 1

```
sig(z) = 1 / 1 + e^(-z)  #Normalization 0..1
```
---

**Adjusting The Weigth of Neuron**:



```
adjustmentOfWeigth = input * error * sigmoid_derivative(output)
```




***Sigmoid Derivative***:
https://towardsdatascience.com/derivative-of-the-sigmoid-function-536880cf918e

formula : ![alt text](https://miro.medium.com/max/1400/1*dKz7wC7Qy17NoG2GVbWvAQ.png)


```
SD = output * (1 - output)   #output: o/p from 1st layer
```
**What Sigmoid Derivatives Does:**
It is used to tunning the weigths based on previous performance

```
If o/p is highly positive / negative:
then only small adjustment needed.


e.g o=0.9 
0.9 * (1 - 0.9) = 0.09
0.1 * (1 - 0.1) = 0.09
```

#Design Custom Neural network


````
import numpy as np
````

````
#to genetrate the random weights between -1..0..1 in columns format
np.random.seed(1) # initailize the random number generator
np.random.random(10) #to generate 10 random values bet 0..1
np.random.random(1) # to generate 1 random value 0..1 (+ve)
np.random.random(1) - 1 # will generate 1 random value -1...0 (-ve)
2*np.random.random() - 1 #will generate 1 random value -1..0..-1 
2 * np.random.random((3,1)) - 1 #will generate 3 rows of random values -1..0..1
````

````
#select dependent and independent variables

X = data[['Wheather', 'Car', 'Friends']] #input
Y = data['Output'] #output

#but these are the data frames to work with numpy methods these needs to be converted into numpy array
X = X.to_numpy()
Y = Y.to_numpy()
Y = Y.reshape((-1,1)) # all varibles are vertical and to make this Y in one column used shape
print(X)
print(Y)
````

````
-0.15778475 + 0.91577906 + 0.06633057
````

````

class MangyaNeuralNetwork():
  def __init__(self):                  #(3,1) == (input layer, output layers)
    self.weights = 2 * np.random.random((3,1)) -1

  #For normalization : sigmoid function ->0..1 Formula sig(z) = 1/1+exp(-z)
  def sigmoid(self , z):
    return 1/(1 + np.exp(-z))

  #sigmoid derivative for weight adjustment formula = SD = o * (1-o)
  def sigmoid_der(self, output):
    return output * (1-output)

  #actual training == calculating accurate weights for the input
  def fit(self,x,y,interations):
    for _ in range(interations):
      #print("weights:",self.weights)




      z = np.dot(x,self.weights)
      #print("Output z:", z)




      output = self.sigmoid(z)
      #print("Sigmoid result:", output)




      error = y - output
     # #print("Error:", error)



      #adjustweight = input * error * SD(output) Deep learning step
      #print("SD",self.sigmoid_der(output))
      #print("err * SD",error * self.sigmoid_der(output))

      #why x need to transpose:> we are only correcting one idependent variables weight at a time 
      #after transpose we will get value related to single column.
      adjustWeight = np.dot(np.transpose(x),(error * self.sigmoid_der(output)))
      #print("weight adjustment", adjustWeight)


      self.weights += adjustWeight

  def predict(self,x):
    z = np.dot(x,self.weights)
    return self.sigmoid(z)

if __name__ == "__main__":
  NNobj = MangyaNeuralNetwork()
  NNobj.fit(X,Y,1000)
  print(NNobj.predict([1,0,1]))
  print(NNobj.weights)
  


  
````
